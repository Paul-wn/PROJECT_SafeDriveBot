```python
from bs4 import BeautifulSoup
import requests
import re
import csv
import pandas as pd
```

```python
url = 'https://www.testdrivinglicence.com/%E0%B8%82%E0%B9%89%E0%B8%AD%E0%B8%AA%E0%B8%AD%E0%B8%9A%E0%B9%83%E0%B8%9A%E0%B8%82%E0%B8%B1%E0%B8%9A%E0%B8%82%E0%B8%B5%E0%B9%88%E0%B8%9E%E0%B8%A3%E0%B9%89%E0%B8%AD%E0%B8%A1%E0%B9%80%E0%B8%89%E0%B8%A5%E0%B8%A2-%E0%B9%80%E0%B8%97%E0%B8%84%E0%B8%99%E0%B8%B4%E0%B8%84%E0%B8%81%E0%B8%B2%E0%B8%A3%E0%B8%82%E0%B8%B1%E0%B8%9A%E0%B8%A3%E0%B8%96%E0%B8%AD%E0%B8%A2%E0%B9%88%E0%B8%B2%E0%B8%87%E0%B8%9B%E0%B8%A5%E0%B8%AD%E0%B8%94%E0%B8%A0%E0%B8%B1%E0%B8%A2/'
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
soup  = BeautifulSoup(response.content , 'html.parser')
contents = soup.find_all('p')
questions = []
for p in contents:
    # ดึงข้อความของ <p> และแยกบรรทัดแรกมาเป็นคำถาม
    question = p.text.strip().split('\n')[0]
    question = re.sub(r'^\d+\.*\s*', '', question)
    questions.append(question)

# แสดงคำถามที่ดึงมาได้
# for q in questions:
#     print(q)


print(len(questions))
print(questions)
```

```python
answers = []
answerss = soup.find_all('strong')
for a in answerss:
    
    text = a.text
    cleaned_text = re.sub(r'^[กขคง]+\.*\s*', '', text)
    answers.append(cleaned_text)

# แสดงคำตอบที่ดึงมาได้
# for ans in answers:
#     print(ans)

print(answers)
```

```python
answers = answers[0:-2:1]
print(answers)
print(len(answers))
```

```python
data = pd.DataFrame({
    'Question': questions,
    'Answer': answers
})

# บันทึกลงในไฟล์ CSV
data.to_csv('Safety_Driving.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

# อ่านไฟล์ CSV
file_path = 'Safety_Driving.csv'
data = pd.read_csv(file_path)

# กรองแถวที่ไม่มีคำว่า "รูป" ในคอลัมน์ Question
filtered_data = data[~data['Question'].str.contains('รูป', na=False)]

# ตรวจสอบข้อมูลที่เหลือหลังการกรอง
print(filtered_data.head())

# บันทึกข้อมูลที่กรองแล้วลงในไฟล์ CSV ใหม่ (ถ้าต้องการ)
filtered_data.to_csv('Safety_Drivingv2.csv', index=False, encoding='utf-8-sig')

print("ข้อมูลที่ไม่มีคำว่า 'รูป' ถูกบันทึกลงในไฟล์ 'Filtered_Safety_Driving.csv' เรียบร้อยแล้ว.")
```
